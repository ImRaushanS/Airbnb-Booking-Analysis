## In the above descriptive analysis of the columns having only numerical values,we found that the min value of price column is zero which is not possible if we came across different values for the same row .Thus,we'll try detecting and tackling outliers from the price column only.

```python
#Price column before removing the outliers
sns.boxplot(ny['price'])
```

```python
#Removing Outliers in Price column
Q1 = np.percentile(df['price'],25)
Q2 = np.percentile(df['price'],50)
Q3 = np.percentile(df['price'],75)
IQR =  Q3 - Q1
a=Q1-(1.5*IQR)
b=Q3+(1.5*IQR)
ny1=ny[(ny.price<b) & (ny.price>a)]
```

```python
#Obtaing the boxplot for the 'Price' column after removal of outliers for it.
sns.boxplot(ny1['price'])
```
